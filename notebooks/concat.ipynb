# Data Final

```python
Bao_final_data = pd.read_csv("D:\\Career\\Research Paper\\MAPR 2026\\ABSA_ITViec_Reviews\\data\\done\\data_Bao_combined.csv")
Huy_final_data = pd.read_csv("D:\\Career\\Research Paper\\MAPR 2026\\ABSA_ITViec_Reviews\\data\\done\\data_Huy_combined.csv")
Long_final_data = pd.read_csv("D:\\Career\\Research Paper\\MAPR 2026\\ABSA_ITViec_Reviews\\data\\done\\data_Long_combined.csv")
```

```python
# Danh sách các cột cần lấy
columns_to_keep = [
    'reviewer_id',
    'timestamp',
    'company_name',
    'company_type',
    'company_size',
    'company_address',
    'review_content'
]

# Concatenate 3 file: data_Bao_combined, data_Huy_combined, data_Long_combined
data_final = pd.concat([
    Bao_final_data[columns_to_keep],
    Huy_final_data[columns_to_keep],
    Long_final_data[columns_to_keep]
], ignore_index=True)

print(f"Số dòng dữ liệu cuối cùng: {data_final.shape[0]}")
print(f"Số cột: {data_final.shape[1]}")
print(f"\nCác cột: {list(data_final.columns)}")
```

```python
# Lưu dữ liệu final
output_path_final = "D:\\Career\\Research Paper\\MAPR 2026\\ABSA_ITViec_Reviews\\data\\done\\data_final.csv"
data_final.to_csv(output_path_final, index=False, encoding='utf-8-sig')

print(f"✓ Dữ liệu final đã được lưu tại: {output_path_final}")
```

```python
# Sắp xếp dữ liệu theo company_name theo thứ tự alphabet
data_final_sorted = data_final.sort_values('company_name', ascending=True).reset_index(drop=True)

print(f"Dữ liệu đã được sắp xếp theo company_name")
print(f"Số dòng: {data_final_sorted.shape[0]}")
```

```python
# Lưu dữ liệu đã sắp xếp
output_path_sorted = "D:\\Career\\Research Paper\\MAPR 2026\\ABSA_ITViec_Reviews\\data\\done\\data_final_sorted.csv"
data_final_sorted.to_csv(output_path_sorted, index=False, encoding='utf-8-sig')

print(f"✓ Dữ liệu đã được sắp xếp và lưu tại: {output_path_sorted}")
```

```python
import pandas as pd
from pathlib import Path

# Load dữ liệu
data_path = Path("../data/final_data_ver2/result_mask_combined.csv")
df = pd.read_csv(data_path, encoding="utf-8-sig")

print("=" * 80)
print("PHÂN TÍCH TỔNG QUAN DATASET")
print("=" * 80)

# Tổng số reviews
total_reviews = len(df)
print(f"\n✓ Tổng cộng Reviews: {total_reviews:,}")

# Số công ty duy nhất
total_companies = df["company_name"].nunique()
print(f"✓ Tổng cộng Công ty: {total_companies}")

# Thống kê bổ sung
print(f"\n--- Thông tin bổ sung ---")
print(f"  • Reviews có masked: {(df['is_masked'] == True).sum():,}")
print(f"  • Reviews không masked: {(df['is_masked'] == False).sum():,}")
print(f"  • Reviews có lỗi chính tả: {(df['spell_has_error'] == True).sum():,}")

# Top 10 công ty có nhiều review nhất
print(f"\n--- Top 10 công ty có nhiều reviews ---")
top_companies = df["company_name"].value_counts().head(10)
for i, (company, count) in enumerate(top_companies.items(), 1):
    print(f"  {i:2d}. {company:<40} {count:>6,} reviews")

print("\n" + "=" * 80)
```
